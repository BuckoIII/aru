We have the following function `transform_input`, which takes as input some data as a string and the transformation to be made, and returns the transformed string.

```python
def transform_input(input_data, transformation):
    if transformation == 'uppercase':
        return input_data.upper()
    if transformation == 'capitalise_first':
        return input_data.capitalize()
    if transformation == 'strip':
        return input_data.strip()
    return input_data

assert transform_input("HeLlO", "uppercase") == "HELLO"
assert transform_input("HeLlO", "capitalise_first") == "Hello"
assert transform_input(" HeLlO ", "strip") == "HeLlO"
```

Re-write this function as `transform_input_strategy` where the second input is now a function to be applied to the input string.

You will need to write functions for each of the transformations you can make.

```python
# Your code here...
def uppercase(string):
    return string.upper()

def capitalise_first(string):
    return string.capitalize()

def strip(string):
    return string.strip()

def transform_input_strategy(string, strategy):
    return strategy(string)
assert transform_input_strategy("HeLlO", uppercase) == "HELLO"
assert transform_input_strategy("HeLlO", capitalise_first) == "Hello"
assert transform_input_strategy(" HeLlO ", strip) == "HeLlO"
```

We now want a new strategy called `strip_metachars` strategy to remove spaces, newline characters ("\n"), and tab characters ("\t").

Create this new strategy and check that it works as expected.

```python
# Your code here...
def strip_metachars(string):
    return string.strip().replace("\t", "")
assert transform_input_strategy("\t HeLlO \n\n", strip_metachars) == "HeLlO"
```
