```python
import requests
from bs4 import BeautifulSoup

class HyperlinkScraper:
    """
    A class that scrapes all hyperlinks at a given URL.
    """

    def __init__(self, url: str):
        self.url = url

    def scrape_hyperlinks(self):
        """
        Scrapes all hyperlinks at the URL.
        """
        response = requests.get(self.url)
        soup = BeautifulSoup(response.text, "html.parser")
        links = []
        for link in soup.find_all("a"):
            href = link.get("href")
            if href and href.startswith("http"):
                links.append(href)
        return links

if __name__ == "__main__":
    scraper = HyperlinkScraper("https://en.wikipedia.org/wiki/English_Civil_War")
    hyperlinks = scraper.scrape_hyperlinks()
    for link in hyperlinks:
        print(link)
```
